```python
import seaborn as sns
import pandas as pd
```

```python
#loading a seaborn built-in dataset
df=sns.load_dataset('tips') #df-dataframe represents the dataset
```

```python
#crosstab gives the summary of the dataset if we want only the data about particular class
pd.crosstab(titanic['Pclass'],titanic['Survived'])
```

A heatmap is a data visualization technique that uses a color-coded grid to represent the magnitude of a variable. It's used to quickly identify patterns, trends, and anomalies within a dataset, especially with a large volume of data.

```python
#heatmap for visualization of the dataset
sns.heatmap(pd.crosstab(titanic['Pclass'],titanic['Survived']))
```

Displot is used to visualize the distribution of a single variable in a dataset. It's a high-level, unified function that can create a histogram, a kernel density estimate (KDE), or an empirical cumulative distribution function (ECDF), making it a flexible and easy-to-use tool for exploring data distributions.



```python
#displot- tells the column from datastet and automatically gives the range or count on another axis
sns.displot(titanic['Age'])
```

```python
#displot of age of people who didnt survived
sns.displot(titanic[titanic['Survived']==0]['Age'])
```

```python
#displot of age count of people who survived
sns.displot(titanic[titanic['Survived']==1]['Age'])
```

A crosstab, or contingency table, is a table used in statistics to analyze the relationship between two or more categorical variables. It displays the frequency distribution of the variables in a matrix format, allowing you to quickly see how the variables are related to each other.

```python
pd.crosstab(titanic['SibSp'],titanic['Survived'])
```

A clustermap is a data visualization that uses a heatmap combined with a dendrogram to display and analyze hierarchical clustering. It's used to identify groups of related rows and columns in a dataset and to visualize the relationships between them.

```python
#clustermap
sns.clustermap(pd.crosstab(titanic['SibSp'],titanic['Survived']))
#lines shown in cluster shows the relation/similar situation and are called tandrums
```

```python
pd.crosstab(titanic['Parch'],titanic['Survived'])
```

```python
sns.clustermap(pd.crosstab(titanic['Parch'],titanic['Survived']))
```

The groupby() method is used for grouping data in a pandas DataFrame. It allows you to split the data into groups based on some criteria (e.g., the values in a specific column). Once the data is grouped, you can apply a function to each group independently.

```python
df2.groupby('year')['passengers'].sum()
#.sum() is use don passengers and the grouping is on year wise
#we are showing yearwise how many total number of passengers traveled
```

```python
new=df2.groupby('year')['passengers'].sum()
#we just assigned the result of this commant to the new datadrame/object i.e how many pasengers in that year stored in it
```

```python
new = df2.groupby('year')['passengers'].sum().reset_index()
#The .reset_index() method then converts the grouped output back into a DataFrame with 'year' and 'passengers' as columns.
sns.lineplot(x='year',y='passengers',data=new)

```
