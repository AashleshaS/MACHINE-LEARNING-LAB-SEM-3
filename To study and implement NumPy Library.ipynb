To Make copy , change the original array, and display both arrays:

```python
arr=np.array([1,2,3,4,5])
x=arr.copy()
arr[0]=42
print(arr)
print(x)

#change wont reflect in copied array(different memory allocation)
```

To Make a view , change the original array,and display both arrays:

```python
ar=np.array([1,2,3,4,5])
x=ar.view()
ar[0]=42
print(ar)
print(x)

#change will reflect in the original array(same memory allocation)
```

To Get Dimensions(Number dimensions) of the array

.ndim is an attribute in NumPy and pandas that returns the number of dimensions of an array or a DataFrame. It is a property, not a function, so you don't use parentheses with it.
* A scalar has 0 dimensions (.ndim returns 0).
* 1D array or a pandas Series has 1 dimension (.ndim returns 1)
* 2D array or a pandas DataFrame has 2 dimensions (.ndim returns 2)
* This attribute is useful for quickly checking the shape and structure of a dataset.

```python
print(a.ndim)
print(b.ndim)
```

**Mathematics/Broadcasting In NumPy**

```python
a=np.array([1,2,3,4])
print(a)

a+=2
print(a)
print(a+2)#add 2 to all elements of array
print(a-2)#subtract 2 from all elements of array
print(a*2)#multipy 2 to all elements of array
print(a/2)#divide all elements of array by 2

b=np.array([1,0,1,0])
print(a+b) # addition of two arrays

print(a**2)#square of all elements of array
```

```python
# to find the sin of all values
print(np.sin(a))
```

```python
# to find the cos of all values
print(np.cos(a))
```

**STATISTICS USING NUMPY**

```python
stats=np.array([[5,2,3],[4,5,6]])
print(stats)
```

```python
print(np.max(stats,axis=1))
#Finds the maximum value in each row (axis=1) and prints a new array containing these maximum values
```

```python
print(np.max(stats,axis=0))
#Finds the maximum value in each column (axis=0) and prints a new array containing these maximum values.
```

```python
print(np.min(stats))
#Finds the minimum value in the entire array and prints it.
```

```python
print(np.sum(stats))
#Calculates the sum of all elements in the array and prints it.
```

```python
sum(sum(stats)) #same result as np.sum(stats) but is less efficient for NumPy arrays.
#Calculates the sum of all elements in the array by summing the sums of each inner list/row.
```
