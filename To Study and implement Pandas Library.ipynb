To read a particular row using indexing

**SORTING OF DATASET**

To sorts the rows based on the value of a particular column in ascending alphabetical order.

```python
df.sort_values('Name')
```

 To do the same sorting but in descending order

```python
df.sort_values('Name',ascending=False)
```

To sort using multiple columns

```python
print(df.sort_values(['Type 1','HP'],ascending=False))
#This code snippet sorts the DataFrame df first by the 'Type 1' column in descending order, and then by the 'HP' column, also in descending order.
```

```python
print(df.sort_values(['Type 1','HP'],ascending=[1,0])) #type1 in ascending and hp in desc.
```
